```python
# 'id' column name of delivery dataset is changed to 'match_id' for easy understanding

delivery.rename(columns = {'id' : 'match_id'}, inplace = True)
```

```python
# Datatypes of 'delivery' dataset

delivery.dtypes
```

```python
# Total runs of the first inning is calculated and 'reset_index()' is used to convert this data into dataframe

total_score_df = delivery.groupby(['match_id', 'inning'], as_index=False)['total_runs'].sum()
total_score_df = total_score_df[total_score_df['inning'] == 1]
total_score_df
```

```python
# Merging the 'match_df' dataframe with the original 'match' dataset

match_df = match.merge(total_score_df[['match_id','total_runs']], left_on='id', right_on='match_id')
match_df
```

```python
match_df.dtypes
```

```python
# Various Teams have played in IPL so far.
# Therefore, some of the teams must be dropped which are not playing in the current season.

match_df['team1'].unique()
```

```python
# These are the teams which are currently playing in IPL.

teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bengaluru',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Lucknow Super Giants',
    'Gujarat Titans'
]
```

```python
# We need to keep the data of those teams only whose name is present in the current teams list.
# We have to discard the other teams' data which are not playing.

match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
# As we can see here, there is the data left only for 909 matches. Earlier it were 1024.

match_df.shape
```

```python
# We will also remove the data of those matches which were affected due to rain.
# We will keep the data of those matches only where 'D/L' method wasn't used.

match_df = match_df[match_df['method'] != 'D/L']
```

```python
# After removing the rain affected matches, we are left with only 892 Matches data.
# So far it has gone: 1024 -> 909 -> 892

match_df.shape
```

```python
# For the final dataset's POV, we need these four columns.

match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
# Now we are merging 'match_df' data with the 'delivery' dataset

delivery_df = match_df.merge(delivery,on='match_id')
```

```python
# We are selecting here the 2nd innings data because our model will predict the probability for the 2nd innings


delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.shape
```

```python
delivery_df.dtypes
```

```python
# We have to create a new column 'current_score' for the 'delivery_df' dataset.
# This column will the tell the current score after every delivery in the match.

# delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
# delivery_df['current_score'] = delivery_df.groupby(['total_runs_y', 'match_id'])['total_runs_y'].cumsum()
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].transform(lambda x: x.cumsum())
delivery_df
```

```python
# We have to calculate the 'runs_left' during the chase now.
# Here '+1' is added as the target is always 1 run more than the 1st Innings score

delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score'] + 1
```

```python
# Now we need to calculate the 'balls_left' in the chase.

delivery_df['balls_left'] = ((20 - (delivery_df['over']+1))*6 + (6 - delivery_df['ball']))
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
# wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
# wickets = delivery_df.groupby(['player_out', 'match_id'])['player_out'].cumsum().values
wickets = delivery_df.groupby('match_id')['player_dismissed'].transform(lambda x: x.cumsum()).values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
# Calculating the current run rate
# crr = runs/overs

delivery_df['crr'] = delivery_df['current_score'] / ( (delivery_df['over']) + (delivery_df['ball']/6) )
delivery_df['crr']
```

```python
# Calculating the required run rate

delivery_df['rrr'] = (delivery_df['runs_left'] / delivery_df['balls_left']) * 6
delivery_df['rrr']
```

```python
delivery_df
```

```python
# Here, we are checking if the chasing team won or not.
# If chasing team won, result is 1, else it is 0.

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

delivery_df['result'] = delivery_df.apply(result,axis=1)
delivery_df
```

```python
# Creating a final DataFrame with all the required columns.

final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
final_df
```

```python
# Some bias might be developed as the balls are in continuation of the same match
# Therefore we need to shuffle the rows

final_df = final_df.sample(final_df.shape[0])
final_df
```

```python
# We are dropping the rows which contains any NULL values

final_df.dropna(inplace=True)
final_df.shape
```

```python
final_df.sample()
```

```python
# For the final time we are checking if there are any NULL Values.

final_df.isnull().sum()
```

```python
# An error was encountered in 'required run rate' and it was showing infinite values for some rows.
# Therefore, in order to resolve this, we are selecting only those rows with non-zero 'balls_left' data

final_df = final_df[final_df['balls_left'] != 0]
final_df.shape
```

```python
# Now we are starting to build the model
# Train test split

X = final_df.iloc[:,:-1] # All the rows except last column
y = final_df.iloc[:,-1] # All the rows with the last column
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
X_train.describe()
```

```python
# Our first 3 columns in above dataframe are string and that would create problem
# Therefore we are applying One Hot encoding over it

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
# This is the code to train the model with Logistic Regression

pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])

# This is the code to train the model with Random Forest Classifier

# pipe = Pipeline(steps=[
#     ('step1',trf),
#     ('step2', RandomForestClassifier())
# ])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
# Our model is giving an accuracy of 80.38% with Logistic Regression
# Logistic Regression internally uses sigmoid function and that would tell the probability of win for both team

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Predicting the probability of both teams

pipe.predict_proba(X_test)[10]
```

```python
# This function will help you to create a dataframe for the 2nd innings of a particular team

def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)

    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]

    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```

```python
# Over by over probability percentage of win and lose of Match ID: 335990

temp_df,target = match_progression(delivery_df,335990,pipe)
temp_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.legend(['Wickets in the over', 'Win Probability', 'Losing Probability', 'Runs scored in the Over'])
# plt.legend('Runs scored in the Over')
plt.title('Target-' + str(target))
```
